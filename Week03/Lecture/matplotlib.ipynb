# Iris Example v1

```python
f = open("iris.txt")

lines = f.readlines()

f.close()

x_values, y_values = [], []
"""
colors = {
    "Iris-versicolor": 0,
    "Iris-setosa": 1,
    "Iris-virginica": 2
}
"""
color = {}
color_codes = []

for line in lines[1:]:
    line = line.strip()
    info = line.split(",")
    sep_len = float(info[1])
    sep_wid = float(info[2])
    species = info[-1]
    if species not in color:
        color[species] = len(color)
    
    x_values.append(sep_len)
    y_values.append(sep_wid)
    color_codes.append(color[species])
    
plt.scatter(x_values, y_values, c=color_codes)
plt.xlabel("Sepal Lenght")
plt.ylabel("Sepal Width")
plt.show()
```

```python
d = {
    "iris_setosa": {
        "x": [2.2, 3.5, 5.6],
        "y": [1.7, 4.2, 4.3]
    },
    "iris_virginica": {
        "x": [2.2, 3.5, 5.6],
        "y": [1.7, 4.2, 4.3]
    },
    "iris_versicolor": {
        "x": [2.2, 3.5, 5.6],
        "y": [1.7, 4.2, 4.3]
    }
}
```

```python
f = open("iris.txt")

lines = f.readlines()

f.close()

d = {}
for line in lines[1:]:
    line = line.strip()
    info = line.split(",")
    sep_len = float(info[1])
    sep_wid = float(info[2])
    species = info[-1]
    if species not in d:
        d[species] = {"x": [], "y": []}
    
    d[species]["x"].append(sep_len)
    d[species]["y"].append(sep_wid)

for k, v in d.items():
    plt.scatter(v["x"], v["y"], label=k)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
plt.show()
```
