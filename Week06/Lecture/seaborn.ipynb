```python
allMeans = []

n_iter = 1000
for i in range(n_iter):
    temp_mean = titanic["fare"].sample(100, replace=True).mean()
    allMeans.append(temp_mean)
```

```python
len(allMeans)
```

```python
allMeans[:5]
```

```python
allMeans.sort()
allMeans[:5]
```

```python
conf_interval = 90
thr = (100 - conf_interval) / 2
idx = int(n_iter * thr / 100)
small = allMeans[idx]
large = allMeans[-idx]
print(small, large)
```

```python
# p1 = (small, 0)
# p2 = (small, 140)
sns.displot(allMeans)
plt.plot([small, small], [0, 140], color="red", linewidth=3, linestyle="dashed")
plt.plot([large, large], [0, 140], color="red", linewidth=3, linestyle="dashed")
plt.show()
```

```python
allMeans = []
conf_interval = 95
n_iter = 1000
for i in range(n_iter):
    temp_mean = titanic["fare"].sample(100, replace=True).mean()
    allMeans.append(temp_mean)

allMeans.sort()
thr = (100 - conf_interval) / 2
idx = int(n_iter * thr / 100)
small = allMeans[idx]
large = allMeans[-idx]
#print(small, large)
sns.displot(allMeans)
plt.plot([small, small], [0, 140], color="red", linewidth=3, linestyle="dashed", label=format(small, ".1f"))
plt.plot([large, large], [0, 140], color="orange", linewidth=3, linestyle="dashed", label=format(large, ".1f"))
plt.legend()
plt.title(str(conf_interval) + "% Confidence Interval")
plt.show()
```
