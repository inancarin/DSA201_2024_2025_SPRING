```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def LinearSearch(value, myList):
    for val in myList:
        if val == value:
            return True
    return False
```

```python
myList = [10, 14, 12, 5, 100]
print(LinearSearch(51, myList))
```

```python
def BinarySearch(val, myList): # assuming myList is sorted
    low, high = 0, len(myList)-1
    while low <= high:
        mid = (low + high) // 2
        if myList[mid] == val:
            return True
        elif myList[mid] < val:
            low = mid + 1
        else:
            high = mid - 1
    return False
```

```python
def BinarySearchRecursive(val, myList, low, high):
    if low > high:
        return False
    mid = (low + high) // 2
    if myList[mid] == val:
        return True
    elif myList[mid] < val:
        low = mid + 1
    else:
        high = mid - 1
    return BinarySearchRecursive(val, myList, low, high)
```

```python
myList = [5, 12, 14, 16, 22, 28, 78, 100]
searchVal = 6
print(BinarySearch(searchVal, myList))
print(BinarySearchRecursive(searchVal, myList, 0, len(myList)-1))
```

```python
for i in range(5, 30):
    n = 2 ** i
    myList = np.arange(1, n, 2) # odd values
    searchVal = 22

    start = time.time()
    res = LinearSearch(searchVal, myList)
    end = time.time()
    diff = end - start
    print("When n is", n, "time is", diff)
```

```python
x_values, y_values_linear, y_values_binary = [], [], []
for i in range(5, 30):
    n = 2 ** i
    myList = np.arange(1, n, 2) # odd values
    searchVal = 22
    x_values.append(n)

    start = time.time()
    res = LinearSearch(searchVal, myList)
    end = time.time()
    diff = end - start
    y_values_linear.append(diff)

    start = time.time()
    res = BinarySearch(searchVal, myList)
    end = time.time()
    diff = end - start
    y_values_binary.append(diff)

plt.plot(x_values, y_values_linear, label="linearSearch")
plt.plot(x_values, y_values_binary, label="binarySearch")
plt.legend()
plt.xlabel("Input Size")
plt.ylabel("# of seconds")
plt.show()
```

```python
def findMaximum(myList):
    if len(myList) == 0:
        return None
    
    maxElem = myList[0]
    for i in range(1, len(myList)):
        if myList[i] > maxElem:
            maxElem = myList[i]
    
    return maxElem
```

```python
for n in [10**8, 2*10**8]:
    myList = np.random.randint(1, 10000000, n)
    start = time.time()
    res = findMaximum(myList)
    end = time.time()
    print("Time:", end-start, "when n is", n)
```

```python
def FindNumberOfUniqueElems(myList): # -> n*n = O(n^2)
    uniqueValues = []
    for elem in myList: # n
        if elem not in uniqueValues: # n
            uniqueValues.append(elem)
    
    return len(uniqueValues)
```

```python
def FindNumberOfUniqueElemswithSet(myList): # O(n)
    mySet = set()
    for elem in myList: # n
        if elem not in mySet: # 1
            mySet.add(elem)
    
    return len(mySet)
```

```python
for n in [20000, 40000]:
    myList = np.random.randint(1, 100000, n)
    start = time.time()
    res = FindNumberOfUniqueElems(myList)
    end = time.time()
    print("Time is", end-start, "when n is", n)
```

```python
for n in [20000, 40000]:
    myList = np.random.randint(1, 100000, n)
    start = time.time()
    res = FindNumberOfUniqueElemswithSet(myList)
    end = time.time()
    print("Time is", end-start, "when n is", n)
```

```python
def Fibonacci(n):
    if n == 1 or n == 0:
        return 1
    return Fibonacci(n-1) + Fibonacci(n-2)
```
